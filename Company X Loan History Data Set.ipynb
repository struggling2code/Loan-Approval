# Project Scenario

Công ty X hoạt động trong tất cả các khoản vay mua nhà. Họ có mặt ở tất cả các khu vực đô thị, nửa đô thị và nông thôn. Khách hàng đầu tiên xin vay mua nhà sau đó công ty xác minh khả năng vay của khách hàng cho một khoản vay.

Công ty muốn tự động hóa quá trình xác minh khả năng vay (trực tuyến) dựa trên thông tin chi tiết của khách hàng được cung cấp trong quá trình điền đơn trực tuyến. Tập dữ liệu này chứa lịch sử vay mua nhà của các khách hàng trước đây của công ty. Chi tiết của mỗi cột trong tập dữ liệu như sau:

* Loan_ID: Định danh duy nhất của khoản vay
* Gender: Nam/Nữ
* Married: Tình trạng hôn nhân của người nộp đơn (Có/Không)
* Dependents: Số lượng người phụ thuộc
* Education: Học vấn của người nộp đơn (Đại học/Trung học)
* Self_Employed: Tự kinh doanh (Có/Không)
* ApplicantIncome: Thu nhập của người nộp đơn
* CoapplicantIncome: Thu nhập của người đồng nộp đơn
* LoanAmount: Số tiền vay trong nghìn đô la
* Loan_Amount_Term: Kỳ hạn của khoản vay theo tháng
* Credit_History: Lịch sử tín dụng đáp ứng các hướng dẫn
* Property_Area: Đô thị / Nửa đô thị / Nông thôn
* Loan_Status: Khoản vay được chấp nhận (Có/Không)

# Literature review

Lin Zhu, Daji Ergu, Kuiyi Liu, Dafeng Qiu và Cai Ying đã đề xuất rằng thuật toán học máy RF (random forest) cho kết quả vượt trội khi so sánh với các thuật toán ML khác bao gồm Decision tree, Logistic regression và SVC (support vector machine) để dự đoán phê duyệt khoản vay. Mô hình của nhóm nghiên cứu cũng hoạt động đối với cả dữ liệu phân loại và dữ liệu số.

Mô hình HMM này đã được train nhiều lần trong giai đoạn train mô hình. Nhóm nghiên cứu đã chỉ ra rằng độ chính xác cao hơn được đạt được bằng cách đào tạo dữ liệu mặc định riêng biệt bằng cách phân đoạn chúng, nếu xác suất cao hơn ngưỡng thì hệ thống giám sát sẽ gửi tín hiệu.

Để dự báo chấp thuận khoản vay, Ashlesha Vaidya đã sử dụng hồi quy logistic như một phương pháp dự đoán và ngẫu nhiên. Tác giả lưu ý rằng vì chúng đơn giản hơn để tạo và cung cấp phân tích dự đoán hiệu quả nhất, mạng nơ-ron nhân tạo và hồi quy logistic đã được sử dụng rộng rãi nhất cho dự đoán khoản vay. Một trong những lý do giải thích cho điều này là các thuật toán khác thường hoạt động kém khi cố gắng dự báo từ dữ liệu không được chuẩn hóa. Tuy nhiên, vì không cần các biến giải thích mà dự báo diễn ra phải có phân phối bình thường, hồi quy logistic có thể dễ dàng xử lý tác động tích cực mạnh và các yếu tố động.

##### Định hình dữ liệu đang bị thiếu

Kiểm tra xem các giá trị bị thiếu liệu ở dạng "NaN" (Not a number) hay ở dạng kí tự đặc biệt nào khác như "?" bằng cách kiểm tra các giá trị xuất hiện trong từng cột của dataframe

```python
columns_to_count = ['Gender', 'Married', 'Dependents', 'Self_Employed', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

for column in columns_to_count:
    column_counts = df[column].value_counts(dropna=False).to_frame(name=f'{column} Counts')
    print(column_counts)
    print("\n" + "="*30 + "\n")  # Dòng phân tách giữa các kết quả
```

Sau khi kiểm tra ta có thể chắc chắn rằng các giá trị bị thiếu đều đồng nhất ở dạng NaN nên ta có thể bắt đầu áp dụng các phương pháp khác nhau để xử lý dữ liệu bị thiếu cho từng cột dữ liệu

**Thay thế dữ liệu bị thiếu bằng giá trị có tần suất xảy ra nhiều nhất**

Cột `Gender` có dữ liệu ở dạng định tính gồm hai giá trị là 'Male' và 'Female', bên cạnh có tới 81% người tới vay theo bộ dữ liệu là 'Male' nên sẽ thay thế 13 giá trị bị thiếu bằng **'Male'**

```python
df['Gender'].replace(np.nan, "Male", inplace = True)
```

Cũng như cột `Gender`, dựa vào tần suất trong dataset nên thay thế giá trị bị thiếu của cột Married bằng giá trị **'Yes'**


```python
df['Married'].replace(np.nan, "Yes", inplace = True)
```

```python
df['Dependents'].replace(np.nan, '0', inplace = True)
```

Vì tần suất xuất hiện giá trị trong cột `Self_Employed` là giá trị 'No' nên thay các giá trị bị thiếu trong cột bằng giá trị **'No'**

```python
df['Self_Employed'].replace(np.nan, "No", inplace = True)
```

Dựa vào tần suất trong dataset, nên thay thế các giá trị bị thiếu trong cột `Credit_History` bằng giá trị **'1.0'**

Các giá trị của nó là các thời gian cố định, được đo bằng số tháng để chỉ kì hạn của khoản vay. Vì vậy, khi làm sạch nó, nhóm không sử dụng bằng giá trị trung bình mà quyết định coi nó là dữ liệu phân loại và thay thế bằng giá trị có tần suất xuất hiện nhiều nhất.

**Thay thế dữ liệu bị thiếu bằng giá trị trung bình**

```python
# Sử dụng boxplot với Seaborn
sns.boxplot(x='Loan_Status', y='LoanAmount', data=df)

# Hiển thị biểu đồ
plt.show()
```

   Quan sát thấy `LoanAmount` có rất nhiều outliers, việc sử dụng giá trị trung bình tức là `mean` để thay thế cho các giá trị bị thiếu có thể dẫn ảnh hưởng trong việc xây dựng và kiểm định mô hình sau. Nhóm sẽ sử dụng `median` để điền vào các giá trị NaN, do đó `mean` không phải là cách tiếp cận đúng vì nó bị ảnh hưởng mạnh bởi sự hiện diện của các giá trị ngoại lệ.

```python
# Tính giá trị trung bình của cột LoanAmount
avg_loan_amount = df["LoanAmount"].astype("float").median(axis=0)
print("Giá trị trung bình của cột LoanAmount:", avg_loan_amount)
```

```python
# Thay thế các giá trị bị thiếu bằng giá trị trung bình của cột LoanAmount
df["LoanAmount"].replace(np.nan, avg_loan_amount, inplace=True)
```

### Ánh xạ các biến phân loại (categorical variables)

Đây là quá trình chuyển đổi các giá trị không liên tục của biến phân loại thành các giá trị có thể sử dụng trong mô hình học máy. Biến phân loại là biến mà giá trị của nó thuộc vào một tập hợp hữu hạn và không liên tục, thường được chia thành các nhóm hoặc danh mục.

```python
mapping_dict = {
    'Gender': {'Male': 0, 'Female': 1},
    'Married': {'No': 0, 'Yes': 1},
    'Education': {'Not Graduate': 0, 'Graduate': 1},
    'Self_Employed': {'No': 0, 'Yes': 1},
    'Credit_History': {1.0: 1, 0.0: 0},
    'Property_Area': {'Urban': 0, 'Rural': 1, 'Semiurban': 2},
    'Dependents': {'0': 0, '1': 1, '2': 2, '3+': 3},
    'Loan_Status': {'N': 0, 'Y': 1}
    }

df.replace(mapping_dict, inplace=True)
```

Ánh xạ này chuyển đổi các giá trị của biến phân loại thành dạng số nguyên để có thể sử dụng trong các mô hình học máy sau này.

**Vậy là chúng ta đã có được một dataset mới 'sạch sẽ' để tiếp tục đi vào quá trình phân tích**

## Khai phá dữ liệu và tìm kiếm insights

Sử dụng heatmap để xem xét sự tương quan tổng quát giữa các biến

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Tạo các khoảng và nhãn cho thu nhập của người đăng ký
bins = [0, 2500, 4000, 6000, 81000]
group_labels = ['Low', 'Average', 'High', 'Very high']

# Tạo một cột mới trong DataFrame df có tên 'Income_bin'.
df['Income_bin'] = pd.cut(df['ApplicantIncome'], bins, labels=group_labels)

# Tạo bảng chéo giữa 'Income_bin' và 'Loan_Status'
income_bin_table = pd.crosstab(df['Income_bin'], df['Loan_Status'])

# Tính tỷ lệ phần trăm và vẽ biểu đồ stacked bar
income_bin_table_percentage = income_bin_table.div(income_bin_table.sum(1).astype(float), axis=0)
income_bin_table_percentage.plot(kind='bar', stacked=True)

# Đặt nhãn cho trục x và trục y
plt.xlabel('ApplicantIncome')
plt.ylabel('Percentage')

# Hiển thị biểu đồ
plt.show()
```

* Binary Logistic Regression: Biến mục tiêu chỉ có hai kết quả có thể xảy ra là Thư rác hoặc Không phải Thư rác, Ung thư hoặc Không Ung thư.

* Multinomial Logistic Regression: Biến mục tiêu có nhiều kết quả có thể xảy ra, ví dụ như dự đoán loại Rượu tương ứng.

Ở đây chúng ta chỉ cần dự đoán xem ngân hàng có chấp thuận cho vay hoặc không cho vay, nên ta sẽ sử dụng Binary Logistic Regression.

Để đánh giá hiệu suất của hồi quy logistic, chúng ta sử dụng các chỉ số sau:

* Accuracy
* Precision
* Recall
* ROC curve
* AUC curve
* Cross-Validation Score
* F1 - score

##### Công thức đánh giá hiệu suất của mô hình hồi quy logistic:

Accuracy (Độ chính xác):
$$\text{Accuracy} = \frac{\text{True Positives} + \text{True Negatives}}{\text{Total Observations}}$$

Precision (Độ chính xác của dự đoán tích cực):
$$\text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$$

Recall (Tỷ lệ bắt giữ được các trường hợp tích cực):
$$\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}$$

$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

$$Precision = \frac{TP}{TP + FP}$$

$$Recall = \frac{TP}{TP + FN}$$

* ROC (Receiver operating characteristic) là một đồ thị được sử dụng khá phổ biến trong validation các model phân loại nhị phân. Đường cong này được tạo ra bằng cách biểu diễn tỷ lệ dự báo true positive rate (TPR) dựa trên tỷ lệ dự báo failse positive rate (FPR) tại các ngưỡng Threshold khác nhau.


* AUC (Area Under the Curve) là một độ đo được sử dụng để đánh giá chất lượng của mô hình phân loại, đặc biệt là khi sử dụng đường cong ROC. AUC có giá trị từ 0 đến 1, với giá trị càng cao càng tốt. Nếu AUC bằng 1, đó chỉ ra một mô hình hoàn hảo, trong khi AUC bằng 0.5 chỉ ra hiệu suất của mô hình là ngẫu nhiên và không có sự phân biệt giữa các lớp. AUC thường được sử dụng để so sánh hiệu suất giữa các mô hình phân loại.

* Cross-Validation Score là một phương pháp đánh giá hiệu suất của mô hình máy học trên dữ liệu huấn luyện. Thay vì chỉ chia dữ liệu thành tập huấn luyện và tập kiểm thử một lần duy nhất, cross-validation thực hiện quá trình này nhiều lần để đánh giá khả năng tổng quát hóa của mô hình.


* F1-score là một metric thường được sử dụng trong bài toán phân loại để đánh giá hiệu suất của mô hình, đặc biệt là trong các tình huống mà sự cân bằng giữa precision và recall là quan trọng.

### Đánh giá mô hình thông qua các chỉ số

* Quan sát Confusion Matrix cho thấy sự chênh lệch lớn giữa True Negative (TN) và False Positive (FP). Sự chênh lệch này có thể gây ra những hậu quả quan trọng trong đánh giá hiệu suất của mô hình phân loại. Nếu chênh lệch này quá lớn, có thể xuất hiện những vấn đề sau. Khi lượng False Positive (FP) tăng lên đáng kể so với True Negative (TN), có thể dẫn đến hiện tượng "False Alarm" (cảnh báo giả mạo). Điều này có nghĩa là mô hình dự đoán quá nhiều trường hợp tích cực (Positive) mà thực tế lại là tiêu cực (Negative). Nếu sự chênh lệch giữa TN và FP quá lớn, độ chính xác của mô hình có thể bị ảnh hưởng mạnh bởi số lượng dự đoán đúng trên tất cả các trường hợp. Điều này có thể dẫn đến việc độ chính xác trở nên không đáng tin cậy, đặc biệt khi dữ liệu mất cân bằng.

**Accuracy**: Độ chính xác tổng thể của mô hình là khoảng 74%, đo lường tỷ lệ các dự đoán đúng (True Positives và True Negatives) trên tổng số mẫu. Trong bối cảnh của bài toán này, độ chính xác là 74%, cho biết tỷ lệ dự đoán đúng trên toàn bộ dữ liệu.

**Precision**: Có nghĩa là độ chính xác, tức là mô hình chính xác đến mức nào. Nói cách khác, khi một mô hình đưa ra dự đoán, mức độ thường xuyên dự đoán đó đúng. Trong bài toán này, khi mô hình Hồi quy Logistic dự đoán người nộp đơn sẽ được vay, thì có tới 75% trường hợp dự đoán chính xác là được vay.

**Recall**: Mô hình chỉ có thể dự đoán chính xác tới 97% số người nộp đơn trong bộ dữ liệu test. Còn 3% còn lại mô hình dự đoán sai khi người nộp đơn không được vay, nhưng thực tế số liệu ghi nhận trong tập test cho rằng người nộp đơn được vay.

* Với giá trị AUC là 0.74, mô hình của có khả năng phân loại tốt hơn so với mô hình ngẫu nhiên (AUC = 0.5), nhưng có thể còn cơ hội để cải thiện.

**Tổng thể, các chỉ số đánh giá hiệu suất của mô hình là khả quan, tuy nhiên, chúng ta cần chú ý đến chỉ số của Confusion Matrix vì nó đóng vai trò quan trọng. Do đó, ta sẽ thực hiện quá trình tinh chỉnh mô hình để đạt được hiệu suất tối ưu nhất có thể.**

Có thể thấy rằng các chỉ số đã được cải thiện một cách đáng kể. Nhóm sẽ tiếp tục tinh chỉnh mô hình bằng cách điều chỉnh ngưỡng `Thresholds`

Tạo và vẽ biểu đồ điểm của độ chính xác và điểm F1 trên cả tập huấn luyện và tập kiểm thử dựa trên các ngưỡng quyết định khác nhau (thresholds) của mô hình Logistic Regression.

```python
#Visualize các ngwo
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Thay thế dữ liệu tương ứng
# X_train: Tập dữ liệu huấn luyện
# y_train: Nhãn của tập dữ liệu huấn luyện
# X_test_imp: Tập dữ liệu kiểm tra hoặc cần dự đoán

train_accuracies = []
train_f1_scores = []
test_accuracies = []
test_f1_scores = []
thresholds = []

for thresh in np.arange(0.1, 0.9, 0.1):
    logreg2_clf = LogisticRegression(solver='liblinear')
    logreg2_clf.fit(X_train, y_train)
    
    y_pred_train_thresh = logreg2_clf.predict_proba(X_train)[:, 1]
    y_pred_train = (y_pred_train_thresh > thresh).astype(int)

    train_acc = accuracy_score(y_train, y_pred_train)
    train_f1 = f1_score(y_train, y_pred_train)
    
    y_pred_test_thresh = logreg2_clf.predict_proba(X_test)[:, 1]
    y_pred_test = (y_pred_test_thresh > thresh).astype(int) 
    
    test_acc = accuracy_score(y_test, y_pred_test)
    test_f1 = f1_score(y_test, y_pred_test)
    
    train_accuracies.append(train_acc)
    train_f1_scores.append(train_f1)
    test_accuracies.append(test_acc)
    test_f1_scores.append(test_f1)
    thresholds.append(thresh)
    
    
Threshold_logreg2 = {
    "Training Accuracy": train_accuracies,
    "Test Accuracy": test_accuracies,
    "Training F1": train_f1_scores,
    "Test F1": test_f1_scores,
    "Decision Threshold": thresholds
}

Threshold_logreg2_df = pd.DataFrame.from_dict(Threshold_logreg2)

plot_df = Threshold_logreg2_df.melt('Decision Threshold', var_name='Metrics', value_name="Values")
fig, ax = plt.subplots(figsize=(15, 5))
sns.pointplot(x="Decision Threshold", y="Values", hue="Metrics", data=plot_df, ax=ax)
plt.show()
```

Điều chỉnh ngưỡng quyết định `Thresholds`

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Huấn luyện mô hình Logistic Regression
logreg2 = LogisticRegression(max_iter=1000)
logreg2.fit(X_train, y_train)

# Dự đoán xác suất cho tập kiểm tra
y_prob = logreg2.predict_proba(X_test)[:, 1]

# Đặt ngưỡng tùy chỉnh 
custom_threshold = 0.70
y_pred_custom_threshold = np.where(y_prob >= custom_threshold, 1, 0)

# Đánh giá mô hình với ngưỡng tùy chỉnh
accuracy_custom_threshold = accuracy_score(y_test, y_pred_custom_threshold)
confusion_matrix_custom_threshold = confusion_matrix(y_test, y_pred_custom_threshold)

print(f"Accuracy with custom threshold ({custom_threshold}): {accuracy_custom_threshold}")
print("Confusion Matrix with custom threshold:")
print(confusion_matrix_custom_threshold)
```
